# Importing Libraries

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing our data

```python
directory = r'data\Bangunan Retak'

classes = os.listdir(directory)

imgs_path = []

imgs_classes = []

for clas in classes:

    full_path = directory + '\\' + clas

    images = os.listdir(full_path)

    images_path = [full_path + '\\' + image for image in images]

    images_class = [clas for img_path in images_path]

    imgs_path.extend(images_path)
    imgs_classes.extend(images_class)
```

```python
df = pd.DataFrame({'path': imgs_path, 'class': imgs_classes})
df['class'].value_counts().plot(kind='bar')
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Distribution of Classes')
plt.xticks(rotation=45)
```

# Exporting our df as a CSV file

```python
df.to_csv('data/images_path.csv', index=False)
```
